# Grade 7 Math Performance Analysis (2019â€“2021)
This notebook explores math performance trends for seventh-grade students over three academic years:
- **2019** (pre-pandemic)
- **2020** (pandemic disruption)
- **2021** (early recovery)

Each year includes a different cohort of 150 students. This is a **cross-sectional** dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

sns.set(style="whitegrid")

# Load data
df = pd.read_csv("C:/Users/heba9/OneDrive/Desktop/class7_math_results.cleaned.csv")
```

##  Boxplot: Math Marks by Year

```python
plt.figure(figsize=(14, 10))
sns.boxplot(
    data=df, x="AcademicYear", y="MathMarks", palette="Set2", width=0.4, linewidth=2
)

plt.title("Math Scores by Academic Year (Grade 7)", fontsize=18)
plt.xlabel("Academic Year", fontsize=14)
plt.ylabel("Math Marks", fontsize=14)
plt.ylim(0, 200)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

##  Distribution: Math Marks by Year

```python
plt.figure(figsize=(12, 6))
sns.histplot(
    data=df,
    x="MathMarks",
    hue="AcademicYear",
    element="step",
    kde=True,
    multiple="stack",
)
plt.title("Distribution of Math Scores by Year")
plt.xlabel("Math Marks")
plt.ylabel("Number of Students")
plt.xlim(0, 200)
plt.tight_layout()
plt.show()
```

##  ANOVA Test: Is There a Significant Difference in Scores?

```python
anova_result = stats.f_oneway(
    df[df["AcademicYear"] == 2019]["MathMarks"],
    df[df["AcademicYear"] == 2020]["MathMarks"],
    df[df["AcademicYear"] == 2021]["MathMarks"],
)
print(f"F-statistic: {anova_result.statistic:.3f}")
print(f"P-value: {anova_result.pvalue:.4f}")
```

### ðŸ” Interpretation
- If the p-value is **less than 0.05**, we reject the null hypothesis and conclude that math performance **varied significantly** across the years.
- If the p-value is **greater than 0.05**, we fail to reject the null hypothesis and conclude there is **no significant difference** in math scores between the years.
